```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

# Load the csv data file to a data frame object 
df = pd.read_csv('death2014.csv')

# Initial exploration of the file by looking at the top several rows
```

```python
sns.set(style="darkgrid")

g = sns.catplot(x="Sex", y="Value", data=df,
                   height=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Deaths")
g.set_xlabels("")
plt.title('Barplot: Deaths by Gender')
plt.show()
```

```python
sns.set(style="whitegrid")

g = sns.catplot(x="Sex", y="Value", data=df,
                   height=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Deaths")
g.set_xlabels("")
plt.title('Pointplot: Deaths y Gender')
plt.show()

```

```python
# Analysis of Countries with the top # of deaths
print("Analysis of countries with the top # of deaths")

# Grouping by Country or Area to get total deaths per country with descending sort
df = df.groupby("Country or Area").Value.sum().reset_index().sort_values("Value", ascending=False)

# Here is a bar graph of the Countries with the top 10 # deaths. We set the label names and title of the graph
df = df.head(10)
df.plot.bar(x="Country or Area", y='Value')
plt.title('Top 10 Countries with Highest # of Deaths in 2014' )
plt.show()
```

```python
plt.scatter(x=df['Country or Area'], y=df['Value'])
plt.show()
```

```python
g = sns.lmplot(y='Value', # Variable 1.
               x='Country or Area', # Variable 2.
               data=df, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("# of Deaths")
g.set_xlabels("Country or Area")
plt.title('Scatterplot: Deaths by Country')
plt.show()
```

```python
g = sns.lmplot(y='Value', 
               x='Value',
               data=df, # Data.
               fit_reg=True, # The regression line also includes a 95% confidence envelope.
               scatter_kws={'alpha':0.4})
g.set_ylabels("# of Deaths")
g.set_xlabels("Country or Area")
plt.title('Scatterplot with regression line: Deaths by Country')
plt.show()
```
